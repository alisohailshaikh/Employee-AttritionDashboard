```python
data['PerformanceRating'] = data['PerformanceRating'].replace({1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'})
```

```python
data['WorkLifeBalance'] = data['WorkLifeBalance'].replace({1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'})
```

```python
import pandas as pd
from scipy.stats import chi2_contingency


contingency_table = pd.crosstab(data['JobSatisfaction'], data['Attrition'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Interpret the results
print(f"Chi2 Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Contingency Table:")
print(expected)
```
